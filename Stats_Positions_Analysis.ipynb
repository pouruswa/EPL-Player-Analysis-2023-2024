```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nation_counts_top10 = player_stats['Nation'].value_counts().head(10)
```

```python
print(plt.colormaps())
```

```python
colors = plt.get_cmap('Set3')(range(10))
plt.figure(figsize=(8, 6))
plt.pie(nation_counts_top10, labels=nation_counts_top10.index, autopct='%1.1f%%', colors=colors)
plt.title('Top 10 Nationalities in EPL 2023-2024 season')
plt.axis('equal')
plt.show()
```

```python
# Correlation Analysis
numerical_columns = player_stats.select_dtypes(include=['float64', 'int64'])
corr_matrix = numerical_columns.corr()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt='.2f', linewidths=0.5, annot_kws={"size": 10}, cbar_kws={"shrink": 0.75})
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```

```python
top_10_goal_scorers = player_stats[['Player', 'Pos', 'Nation', 'Squad', 'Perf_Gls']].sort_values(by='Perf_Gls', ascending=False).head(10)
print("Top 10 Goal Scorers in EPL:")
print(top_10_goal_scorers)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(player_stats['Age'], player_stats['Perf_Gls'], s=100, color='skyblue', alpha=0.7)
top_scorers = player_stats[player_stats['Perf_Gls'] > 15]
for _, player in top_scorers.iterrows():
    plt.text(player['Age'] + 0.2, player['Perf_Gls'] + 0.2,
             player['Player'], fontsize=9, color='darkblue')
plt.title('Top Scorers across Age Groups')
plt.xlabel('Age')
plt.ylabel('Number of Goals Scored')
plt.grid(True)
plt.show()
```

```python
top_assisting_players = player_stats.sort_values(by='Perf_Ast', ascending=False).head(10) #overall performance assists
position_assists = player_stats.groupby('Pos')['Perf_Ast'].sum().sort_values(ascending=False)
```

```python
top_assisting_players_90 = player_stats.sort_values(by='90m_Ast', ascending=False).head(10) #90 minutes assists
top_10_assisters_90 = player_stats[['Player', 'Nation', 'Pos', 'Squad', '90m_Ast']].sort_values(by='90m_Ast', ascending=False).head(10)
print("Top 10 Assisters in EPL for 90 minutes played:")
print(top_10_assisters_90)
```

```python
top_10_assisters = player_stats[['Player', 'Nation', 'Pos', 'Squad', 'Perf_Ast']].sort_values(by='Perf_Ast', ascending=False).head(10)
print("Top 10 Assisters in EPL:")
print(top_10_assisters)
```

```python
plt.figure(figsize=(12, 6))
plt.barh(position_assists.index, position_assists.values, color='lightgreen')
plt.xlabel('Position')
plt.ylabel('Total Assists')
plt.title('Total Assists by Position in EPL 2023-2024 Season')
plt.gca().invert_yaxis()
plt.grid(False)
```

```python
top_defenders_assists = player_stats[player_stats['Pos'] == 'DF'].sort_values(by='Perf_Ast', ascending=False).head(10)
top_midfielders_assists = player_stats[player_stats['Pos'] == 'MF'].sort_values(by='Perf_Ast', ascending=False).head(10)
```

```python
summary_stats = player_stats[['PrgPass', 'PrgCarries', 'PrgPass_Rec']].describe()
min_values = summary_stats.loc['min']
max_values = summary_stats.loc['max']
print("Minimum values:")
print(min_values)
print("\nMaximum values:")
print(max_values)
```

```python
player_stats['PrgPass_norm'] = (player_stats['PrgPass'] - player_stats['PrgPass'].min()) / (player_stats['PrgPass'].max() - player_stats['PrgPass'].min()) * 100
player_stats['PrgCarries_norm'] = (player_stats['PrgCarries'] - player_stats['PrgCarries'].min()) / (player_stats['PrgCarries'].max() - player_stats['PrgCarries'].min()) * 100
player_stats['PrgPass_Rec_norm'] = (player_stats['PrgPass_Rec'] - player_stats['PrgPass_Rec'].min()) / (player_stats['PrgPass_Rec'].max() - player_stats['PrgPass_Rec'].min()) * 100
player_stats['playmaking_ability'] = (player_stats['PrgPass_norm'] * 0.4) + (player_stats['PrgCarries_norm'] * 0.4) + (player_stats['PrgPass_Rec_norm'] * 0.2)
player_stats['playmaking_ability'] = player_stats['playmaking_ability'].round(2)
player_stats[['Player', 'playmaking_ability']].head()
top_playmakers_10 = player_stats.sort_values(by='playmaking_ability', ascending=False).head(10)
top_playmakers_10[['Player', 'Squad', 'playmaking_ability']]
```
